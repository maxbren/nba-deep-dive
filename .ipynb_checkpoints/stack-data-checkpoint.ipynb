```python
import pandas as pd
```

```python
seasons = ['19-20', '18-19', '17-18', '16-17', '15-16']
table_labels = ['player-stats', 'team-stats', 'games']
new_tables = {}
```

```python
for label in table_labels:
    tables = []
    for season in seasons:
        filepath = 'processed-data/{}/{}.xlsx'.format(season, label)
        df = pd.read_excel(filepath)
        df['Season'] = [season] * len(df)
        cols = df.columns.tolist()
        new_cols = [cols[0]] + [cols[-1]] + cols[1:-1]
        df = df[new_cols]
        tables.append(df)
    new_tables[label] = pd.concat(tables)
    new_tables[label].reset_index(inplace=True, drop=True)
    new_tables[label].to_excel('processed-data/stacked-{}.xlsx'.format(label), index=False)
```
