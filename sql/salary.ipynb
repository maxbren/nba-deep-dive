```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('salary-player-stats.db')

df = pd.read_excel('../processed-data/salary.xlsx', sheet_name='salary')
df.to_sql('salary', conn, index=False,if_exists="replace")

df = pd.read_excel('../processed-data/stacked-player-stats.xlsx', sheet_name='player-stats')
df.to_sql('players', conn, index=False,if_exists="replace")

curr = conn.cursor()
```

```python
def query_db(query):
    execute = curr.execute(query)
    cols = [column[0] for column in execute.description]
    results = pd.DataFrame.from_records(data=execute.fetchall(), columns=cols)
    return results
```

```python
# Salary per Position
query = \
'''
SELECT Pos, AVG(Amount), MIN(Amount), MAX(Amount)
FROM salary
JOIN players ON salary.player is players.player
WHERE Pos in ("C", "SF", "PF", "SG", "PG") AND season is "17-18"
GROUP BY Pos
ORDER BY AVG(Amount) DESC
'''
query_db(query)
```

```python
# Difference between Offense and Defense (Season 19-20)
query = \
'''
SELECT Player, (OWS - DWS) as "Off vs Def"
FROM players
WHERE Season is "19-20"
ORDER BY "Off vs Def" DESC
'''
query_db(query)
```

```python
# Difference between Offense and Defense (Season 15-16)
query = \
'''
SELECT Player, (OWS - DWS) as "Off vs Def"
FROM players
WHERE Season is "15-16"
ORDER BY "Off vs Def" DESC
'''
query_db(query)
```
